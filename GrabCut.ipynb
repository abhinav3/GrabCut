# GrabCut Preliminary experiment

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('2.jpg')
mask = np.zeros(img.shape[:2],np.uint8)
# Since OpenCV represents RGB images as multi-dimensional NumPy arrays…but in reverse order!
# This means that images are actually represented in BGR order rather than RGB!
# All we need to do is convert the image from BGR to RGB:
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.colorbar(),plt.show();
```

```python
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
```

```python
rect = (10,10,700,700) #to decide on 1st and 2nd arg
cv2.grabCut(img,mask,rect,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT); 
```

```python
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
img = img*mask2[:,:,np.newaxis]
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.colorbar(),plt.show();
```
